## 1. Product

```python
top = ['반팔', '나시', '가디건']

# 3일동안 입을 상의의 경우의 수?

for x in top:
    for y in top:
        for z in top:
            print(x, y, z)
```

```python
from itertools import product

for p in product(top, top, top):
    print(p)
```

```python
top = ['반팔', '나시', '가디건']
bottom = ['청바지', '면바지']
shoes = ['슬리퍼', '운동화', '장화']

# 가능한 코디의 경우의 수?

from itertools import product

for p in product(top, bottom, shoes):
    print(p)
```

## 2. Permutation

```python
top = ['반팔1', '반팔2', '반팔3', '나시', '가디건1', '가디건2']

# 월 화 수 목 금 겹치지 않게 입으려 할때 경우의 수?

from itertools import permutations

for p in permutations(top, 5):
    print(p)
```

## 3. Combinations

```python
# 옷을 3개 살 때 경우의 수?

from itertools import combinations

top = ['반팔1', '반팔2', '반팔3', '나시', '가디건1', '가디건2']

for c in combinations(top, 3):
    print(c)
```

## 4. Combinations with Replacement

```python
# 같은 옷을 여러개 사도 된다. 5개를 살때 경우의 수는?

from itertools import combinations_with_replacement

top = ['반팔1', '반팔2', '반팔3']

for c in combinations_with_replacement(top, 5):
    print(c)
```
